Trying out parameterized types over `typing.Self` for `To...` types and `of` class methods.

```python
from typing import Self
from splatlog.json import JSONFormatter, ToJSONFormatter


class ExtFormatter(JSONFormatter):
    @classmethod
    def of(cls: type[Self], value: ToJSONFormatter[Self]) -> Self:
        if isinstance(value, cls):
            return value
        return super().of(value)


# Ok called with instance of invoking class
x = ExtFormatter()
assert x is ExtFormatter.of(x)

# Type error, because invoked through `ExtFormatter` must return `ExtFormatter`
ExtFormatter.of(JSONFormatter())
```
