```python
import dataclasses as dc

from rich import measure as measure
from rich.segment import Segment
from rich.console import Group
from rich.markdown import Markdown


import splatlog as slog
from splatlog.rich.section import Heading, Section
from splatlog.rich.framing import with_framing

slog.setup(level=slog.INFO, console=True, theme=slog.rich.THEME_ANSI_DARK)

LOG = slog.getLogger(__name__)

c = slog.rich.to_console({"width": 80}, theme=slog.rich.THEME_ANSI_DARK)
```

```python
top = Heading("Error", style="traceback.title", level=1, align="center")

sections = [
    Section(
        title=title,
        style=style,
        level=2,
        content=Markdown(f"Content for _{title}_ section."),
        padding=(0, 0, 1, 0),
    )
    for title, style in (
        ("Invocation", "cyan"),
        ("Exception", "red"),
        ("Help", "green"),
    )
]

c.print(Group(top, *sections))
```
