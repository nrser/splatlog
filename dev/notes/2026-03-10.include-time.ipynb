Include Time in `RichHandler` Output
==============================================================================

In the `RichHandler` output. I know it worked at some point so far back, and I
finally have another need for it.

```python
import time
from typing import cast
import splatlog as slog

slog.setup(
    level=slog.DEBUG,
    console={
        "show_path": False,
        "time": True,
    },
    theme=slog.rich.THEME_ANSI_DARK,
)
log = slog.getLogger("splatlog.rich.RichHandler")

log.debug("Hey!")

t = 0.5
time.sleep(t)

log.debug(
    "Hey!",
    slept_for=t,
    time_config=cast(
        slog.rich.RichHandler, slog.named_handlers.get("console")
    ).time_config,
)
```
