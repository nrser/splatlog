```python
from collections.abc import Iterable as Iterable
import sys

from rich.style import Style as Style
from rich.console import Console

import splatlog as slog
```

```python
slog.setup(
    level=slog.INFO,
    console={"console": {"file": sys.stdout, "width": 120}},
    theme=slog.rich.THEME_ANSI_DARK,
)
```

```python
def get_syntax_error() -> SyntaxError:
    try:
        import tmp  # noqa: F401
    except SyntaxError as syn_err:
        return syn_err
    raise Exception("unreachable")


syn_err = get_syntax_error()
syn_err
```

```python
console = Console(width=80)
enriched_syn_err = slog.rich.EnrichedException(
    syn_err, width=80, syntax_theme="nord-darker"
)
```

```python
console.print(enriched_syn_err)
```

```python
with console.use_theme(enriched_syn_err.theme):
    console.print(enriched_syn_err)
```
