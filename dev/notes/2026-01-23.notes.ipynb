```python
from rich.console import Console
from rich.style import Style
from rich.text import Text

import splatlog

theme = splatlog.rich.override_ansi_colors(
    splatlog.rich.THEME, **splatlog.rich.ANSI_PALETTE_DARK
)
c = Console(theme=theme)

c.print("[red]hot[/] [log.name]name[/] [log.class]Class[/]")

name = "splatlog.splat_logger.ClassLogger"
class_name = "ClassLogger"

t_1 = Text()
t_1.append(name, style="log.name")

c.print(t_1)

c.print(t_1.spans)

t_1.stylize("log.class", -1 * len(class_name))
t_1.highlight_words((".",), style="log.name.sep")

c.print(t_1)
c.print(t_1.spans)
```

```python
def find_all(s, char):
    i = s.find(char)
    while i != -1:
        yield i
        i = s.find(char, i + 1)


t_2 = Text()


t_2.append(name)

start = 0
for i in find_all(name, "."):
    t_2.stylize("log.name", start, i)
    start = i + 1
    t_2.stylize("log.name.sep", i, start)

style = "log.class" if name.find(class_name, start) > 0 else "log.name"
t_2.stylize(style, start)


c.print(t_2)
c.print(t_2.spans)
```
